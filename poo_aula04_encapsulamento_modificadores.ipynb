```python
class Pessoa:
    ano_atual = 2026

    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def get_ano_nascimento(self):
        return self.ano_atual - self.idade
```

```python
Pessoa.ano_atual
```

```python
julles = Pessoa('Julles', 30)
julles.get_ano_nascimento()
```
